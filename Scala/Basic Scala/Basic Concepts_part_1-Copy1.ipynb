#### The Sustitution model

Evaluation strategy:
- Call by value: has the advantage that it evalues every function argument only once.
- Call by name: has the advantage that a function argument is not evaluated if the corresponding parameter is unused in the evaluation of the function body

...both strategies reduce to the final values.

If CBV evaluation of a expression 'e' terminates, then CBN evaluation of 'e' terminates, too. The other direction is not true.

To use call by name, we need to indicate it with =>

```scala
def constOne(x: Int, y: => Int) = 1
```

```scala
constOne(1+2, loop)  // the second argument is not used...
```

but ... it fails because evaluates the second parameter :)

```scala
def constOne(x: Int, y: Int) = 1
constOne(1+2, loop)
```
